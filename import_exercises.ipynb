 2.1  * How many different ways can you combine the letters from "abc" with the numbers 1, 2, and 3?

```python
import itertools
itertools.product("abc","123")
# we need a lis so with list we can get this:
list(itertools.product("abc","123"))
#How many different ways
len(list(itertools.product("abc","123")))
```

2.2 How many different combinations are there of 2 letters from "abcd"?

```python
from itertools import combinations
x=0
for c in itertools.combinations('abcd',2):
    x+=1
    print(c)
print("differents combinations of two letters:", x)
```

```python
#other way
itertools.combinations('abcd',2)
list(itertools.combinations('abcd',2))
len(list(itertools.combinations('abcd',2)))
```

2.3 How many different permutations are there of 2 letters from "abcd"?

```python
from itertools import permutations
x=0
for p in itertools.permutations ("abcd",2):
    x+=1
    print(p)
print("differents permutation of 2 letters:", x)
    
```

```python
#other way
itertools.permutations('abcd',2)
list(itertools.permutations('abcd',2))
len(list(itertools.permutations('abcd',2)))
```

```python
#we are counting how many for each fruit
fruits_counts = {}
for fruit in fruit_list:
    if fruit not in fruits_counts.keys():
        fruits_counts[fruit] = 1
    else:
        fruits_counts[fruit] +=1
fruits_counts
```

```python
#les see .items
for k,v in fruits_counts.items():
    print(k)
    print(v)
```
